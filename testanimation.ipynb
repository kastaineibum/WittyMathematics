```python
from matplotlib import animation
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import MultipleLocator

fig,pltsub=plt.subplots()

x1=np.linspace(-10,10,50) #替换这里的横轴显示范围和取样点个数
y1=x1 #替换这里的函数表达式
x2=np.linspace(-10,10,50) #替换这里的横轴显示范围和取样点个数
y2=x2+3 #替换这里的函数表达式
x3=np.linspace(-10,10,50) #替换这里的横轴显示范围和取样点个数
y3=2*x3+3 #替换这里的函数表达式

pltsub.axis('equal')
line1,=pltsub.plot(x1, y1)
line2,=pltsub.plot(x2, y2)
line3,=pltsub.plot(x3, y3)
x_major_locator=MultipleLocator(1) #把x轴的刻度间隔设置为1，并存在变量里
y_major_locator=MultipleLocator(1) #把y轴的刻度间隔设置为1，并存在变量里
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)#把x轴的主刻度设置为1的倍数
ax.yaxis.set_major_locator(y_major_locator)#把y轴的主刻度设置为1的倍数
plt.xlim(-10,10) #x轴的刻度范围设置
plt.ylim(-10,10) #y轴的刻度范围设置
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
 
def update(i):
    line1.set_ydata(i*x1)
    return line1,
def init():
    line1.set_ydata(x1)
    return line1,
 
ani=animation.FuncAnimation(fig=fig,func=update,frames=100,init_func=init,interval=10,blit=False)
ani.save('test_animation.gif',writer='imagemagick')
plt.show()
```
